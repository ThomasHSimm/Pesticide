```python
column_names_dict={'amount_detected':  'Pesticide residues \nfound (mg/kg)' ,
                    'mrl': 'Maximum reporting \nlimit MRL (mg/kg)', 
                    'amount_pc' : 'Pesticide residues \nfraction of MRL'}



def range_plots(df2, plot_type='boxplot'):
    
    numeric_df = df2.loc[:,['amount_detected','mrl']].copy()
    numeric_df = numeric_df.loc[ numeric_df['amount_detected']!=0]
    
    numeric_df['amount_pc']=numeric_df['amount_detected']/numeric_df['mrl']
    numeric_df = numeric_df.loc[ numeric_df['amount_pc'].notnull()]
    numeric_df = numeric_df.rename(columns = column_names_dict)

    if plot_type=='boxplot':
        fig = plt.figure(figsize=(8,8))
        sns.boxplot( data = numeric_df )
        plt.yscale('log')
    elif plot_type=='hist':
        fig, ax = plt.subplots(nrows=1, ncols=3, sharex=False,
                                    figsize=(12, 6))

        max_bin = [5, 20, 1]
        for i,cols in enumerate(numeric_df):
            data_col = numeric_df[cols]
            counts, bins = np.histogram(data_col,
                    range = [0,max_bin[i]],bins=30)
            ax[i].hist(bins[:-1], bins, weights=counts)
            # ax[i].set_yscale('log')
            ax[i].set_title(cols)
            if i==0:
                ax[i].set_ylabel('Count')
    
    # return fig

range_plots(df2)
range_plots(df2,'hist')
```

```python
# df2_grouped = df2.groupby('country_of_origin',as_index=False).mean(numeric_only =True)
# df2_grouped_b = df2.groupby('country_of_origin', as_index=False).count().iloc[:,0:2]

# df2_grouped= df2_grouped.merge(df2_grouped_b, left_on='country_of_origin', right_on='country_of_origin')
# df2_grouped.rename(columns ={'sample_id':'Number of tests'},inplace=True)
# df2_grouped = df2_grouped.rename(columns = column_names_dict)
# df2_grouped= df2_grouped.sort_values('Number of tests', ascending=False)

def groupby_id_country(df2: pd.DataFrame) -> pd.DataFrame:
    """
    Groups a Pandas DataFrame based on the sample_id and country_of_origin
        the new dataframe has a new column number_of_tests
        - this is the number of unique sample_ids
        the other 2 numerical columns are means of previous values

    Args:
        df2 (pd.DataFrame): Pandas DataFrame. Pesticide data after 1st clean

    Raises:
        ValueError: ??

    Returns:
        df2_grouped (pd.DataFrame): Pandas DataFrame grouped by id and country- note the mean is taken twice
                                        1. When grouping by id
                                        2. When grouping by country
                                    the new dataframe has a new column number_of_tests
                                        - this is the number of unique sample_ids
                                    in addition to the previous numeric columns
        df2_grouped_sample pd.DataFrame): Pandas DataFrame grouped by id only
    """
    # group by id
    df2_grouped_sample = df2.groupby(['sample_id','date_of_sampling','country_of_origin'],as_index=False).mean(numeric_only =True).sort_values('amount_detected', ascending=False)
    
    # group by country-> mean
    df2_grouped = df2_grouped_sample.groupby('country_of_origin',as_index=False).mean(numeric_only =True)
    # group by country-> count
    df2_grouped_b = df2_grouped_sample.groupby('country_of_origin', as_index=False).count().iloc[:,0:2]
    # merge the 2 new dfs and rename count column
    df2_grouped= df2_grouped.merge(df2_grouped_b, left_on='country_of_origin', right_on='country_of_origin')
    df2_grouped.rename(columns ={'sample_id':'number_of_tests'},inplace=True)
    # sort dataframe by counts
    df2_grouped= df2_grouped.sort_values('number_of_tests', ascending=False)

    df2_grouped.reset_index(inplace=True, drop=True)
    df2_grouped_sample.reset_index(inplace=True, drop=True)

    return df2_grouped, df2_grouped_sample

df2_grouped, _ = groupby_id_country(df2)
df2_grouped
```
